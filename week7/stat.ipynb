# 2. Categorizing Data into Categorical & Numerical

Loan_ID - nominal 
Even though it has numbers in them its not measurable or countable

Gender - nominal (label with no order)

Married - nominal (label with no order)

Dependents - ordinal (has ordering) : not numerical because it has 3+ and not 3

Education - nominal (label with no order)

Self_Employed - nominal (label with no order)

ApplicantIncome - continuous (they can take any value)

CoapplicantIncome - continuous (any values has decimals too)

LoanAmount - continuous (any value)

Loan_Amount_term - discrete(measure taken in steps)

Credit_History - discrete

Property_Area - ordinal (ordering is present) : urban,rural,semiurban

Loan_Status - nominal (label with no order)

```python
df['Loan_ID'].isnull().sum()
```

```python
len(df['Loan_ID'].unique())
```

```python
nom_col = ['Gender', 'Married', 'Self_Employed', 'Credit_History']
ord_col = ['Dependents']
dis_col = ['Loan_Amount_Term']
```

```python
df['Dependents'].isnull().sum()
sns.countplot(x='Dependents',data=df,hue='Loan_Status (Approved)')
```

```python
df[nom_col] = df[nom_col].fillna(df[nom_col].mode().iloc[0])
```

```python
df[ord_col] = df[ord_col].fillna(df[ord_col].mode().iloc[0])
```

```python
df['ApplicantIncome'].isnull().sum()
```

```python
df['CoapplicantIncome'].isnull().sum()
```

```python
df['LoanAmount'].isnull().sum()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
print(df['LoanAmount'].isnull().sum())
```

```python
df['Loan_Amount_Term'].isnull().sum()
```

```python
df[dis_col]=df[dis_col].fillna(df[dis_col].mean())
print(df['Loan_Amount_Term'].isnull().sum())
```
